```python
from alchemyapi import AlchemyAPI
import json
```

```python
alchemyapi = AlchemyAPI()
```

```python
from pymongo import MongoClient

client = MongoClient()
db = client.alchemy

db.drop_collection("entities")
```

```python
urls = ["http://www.huffingtonpost.com/2015/05/04/harry-reid-tpp_n_7208076.html",
"http://www.huffingtonpost.com/2015/05/05/mike-huckabee-2016_n_7120826.html",
"http://www.nytimes.com/2014/03/17/us/politics/despite-leaps-women-lag-in-politics.html?_r=0",
"http://www.huffingtonpost.com/2015/02/03/barack-obama-mitt-romney_n_6609530.html"]
```

```python
# May want to keep entire json structure so I know the original url used
for u in urls:
    response = alchemyapi.entities('url',u, {'knowledgeGraph': 1, 'sentiment': 1})
    if response["status"] == "OK":
        db.entities.insert_many(response["entities"])
```

```python
cursor = db.entities.aggregate(
    [
        {"$group": {"_id": "$type", "count": {"$sum": 1}}}
    ]
)

for doc in cursor:
    print doc
```

```python
cursor = db.entities.find({"type": "Person"})

for doc in cursor:
    print doc
```

```python
cursor = db.entities.aggregate(
    [
        {"$match": {"type": "Person"}},
        {"$group": {"_id": "$text", "count": {"$sum": 1}, "totalSentiment": {"$sum": "$sentiment.score"}}}
    ]
)

for doc in cursor:
    print doc
```
